### **(a)  Write a program with Matplotlib to visualize scatter plot below.**

Input

```python
# Values of coordinate x
x = np.arange(0, 5, 0.2)

# Values of coordinate y1 (linear line)
linearFunction = np.vectorize(lambda x : x + 0.25)
y1 = linearFunction(x)

# Values of coordinate y2 (parabolic 1)
parabolicFunction1 = np.vectorize(lambda x : x ** 2)
y2 = parabolicFunction1(x)

# Values of coordinate y3 (parabolic 2)
parabolicFunction2 = np.vectorize(lambda x : 2 * (x ** 2))
y3 = parabolicFunction2(x)
```

Output

```python
for y, str_y, clr in zip([y1, y2, y3], ['y1', 'y2', 'y3'], ['r', 'g', 'b']):
    plt.scatter(x, y, label = str_y, 
                color = clr, 
                linewidths = 2)
plt.legend()
plt.show()
```

```python
# Dataset Merge

merge_result_1_df = pd.merge(df_shs, df_teachers, 
                             on = ['id'], 
                             suffixes = ('', '_remove'), 
                             how = "inner")
final_high_school_df = pd.merge(merge_result_1_df, df_students, 
                                on = ['id'], 
                                suffixes = ('', '_remove'), 
                                how = "inner")

# Display New Dataframe 5 Top Rows
display(final_high_school_df.head())

# Droping columns with value suffix "remove"
final_high_school_df.drop([i for i in final_high_school_df.columns if 'remove' in i], 
                          axis = 1, 
                          inplace = True)

display(final_high_school_df.head())
final_high_school_df.info()
```

```python
if final_high_school_df.isna().sum().sum() == 0:
    print('The new dataframe does not have any missing values.')
else:
    print(f'There are {final_high_school_df.isna().sum().sum()} missing values')
    for col in final_high_school_df:
        if final_high_school_df[col].dtype == 'object':
            final_high_school_df[col].fillna(final_high_school_df[col].mode()[0], inplace=True)
        else:
            final_high_school_df[col].fillna(0, inplace=True)
```

## 4  Hotel Booking

---

There is Jakarta hotel booking data (hotel_booking.csv provided). Use data only arrival_date_year in
2016. Then do the following processing.

### **a. Read the CSV file and get the average number of bookings each month and visualize it in Barchart**

```python
df_hotel = pd.read_csv('data/hotel_bookings.csv')
display(df_hotel.head())
df_hotel.info()
```

```python
df_hotel_2016 = df_hotel[df_hotel.arrival_date_year == 2016].reset_index(drop=True)

avg_book_permonth = pd.pivot_table(data=df_hotel_2016,
                                   values='hotel',
                                   index='arrival_date_month',
                                   aggfunc='count')

month_order = ['January', 'February', 'March', 'April', 
               'May', 'June', 'July', 'August', 
               'September', 'October', 'November', 'December']


avg_book_permonth.index = pd.CategoricalIndex(avg_book_permonth.index, categories=month_order)
avg_book_permonth.sort_index(inplace=True)

fig = go.Figure(go.Bar(
    x=avg_book_permonth.index,
    y=avg_book_permonth['hotel'], 
))

fig.update_layout(
    title='Average Number of Bookings Per Month in 2016',
    xaxis_title='Month',
    yaxis_title='Number of Bookings',
    height=500,
    template='plotly', 
)

fig.show()
```

### **b. Visualize the number of adult guests into diagram / chart that you think is appropriate**

```python
count_adult_guests = pd.pivot_table(data=df_hotel_2016,
                                    values='adults',
                                    index='hotel',
                                    aggfunc='sum')

fig = go.Figure(go.Pie(
    labels=count_adult_guests.index,  
    values=count_adult_guests['adults'],  
    sort=True 
))

fig.update_layout(
    height=400, 
    width=400
)

fig.show()
```

### **c. Visualize the number of canceled bookings (is_cancelled, 0 = not cancelled, 1 = canceled) each week using the appropriate chart**

```python
canceled_books_weekly = pd.pivot_table(data=df_hotel_2016[df_hotel_2016.is_canceled == 1],
                                       values=['is_canceled'],
                                       columns=['arrival_date_week_number'],
                                       aggfunc='count').T.rename({'is_canceled': 'cancel_bookings'}, axis=1)


fig = go.Figure()

fig.add_trace(go.Scatter(
    x=canceled_books_weekly.index,  
    y=canceled_books_weekly['cancel_bookings'],  
    mode='lines', 
    name='Canceled Bookings', 
))

fig.update_layout(
    title='Number of Canceled Bookings Each Week',
    xaxis_title='Week',
    yaxis_title='Number of Canceled Bookings',
    title_font_size=20,
    height=500, 
    template='plotly',  
    showlegend=False,  
)

fig.show()
```
