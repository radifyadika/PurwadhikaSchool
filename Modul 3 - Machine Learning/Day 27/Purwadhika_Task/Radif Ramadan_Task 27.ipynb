```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

1. Buatlah Dataframe yg berisi List digimon yg berasal dari web `http://digidb.io/digimon-list/` .

```python
df = pd.read_html('http://digidb.io/digimon-list/')
df
```

```python
df = df[0]
df
```

2.  Akses Digimon yg bertipe Vaccine dan memiliki HP diatas 1100.

```python
df[(df.Type == 'Vaccine') & (df.HP > 1100)]
```

3.  Berapa jumlah Digimon Virus yg memiliki HP diatas rata-rata, memiliki Memory diatas 18,
memiliki Attack diatas 100, dan speed nya diatas atau sama dengan rata-rata. Output berupa
Nilai (Angka Jumlahnya)

```python
digimon_virus_stat = df[
    (df.Type == 'Virus') & 
    (df.HP > df.HP.mean()) & 
    (df.Memory > 18) & 
    (df.Atk > 100) & 
    (df.Spd >= df.Spd.mean())
]

print(f'Jumlah Digimon Virus dengan Stats terkait : {len(digimon_virus_stat)}')
```

```python
digimon_virus_stat
```

4. Simpan dataframe tadi ke dalam bentuk file csv dg nama Digimon.csv, tanpa menggunakan
Index

```python
digimon_virus_stat.to_csv('digimon.csv', index = False)
```

5.  Buat table yg menunjukan Jumlah Digimon dari setiap Type Digimon (DataFrame).

```python
df.groupby("Type")[["Digimon"]].count()
```

6. Buat table yg menunjukkan rata-rata HP dan Attack tertinggi Berdasarkan Stage Digimonnya.

```python
df.groupby("Stage")[["HP", "Atk"]].mean().round(0)
```

7.  Ubah nilai HP Kuramon menjadi 1000 dan Attack nya menjadi 2500.

```python
df.loc[df.Digimon == 'Kuramon', 'HP'] = 1000
df.loc[df.Digimon == 'Kuramon', 'Atk'] = 2500
df[df.Digimon == 'Kuramon']
```

8.  Tampilkan 3 Stage Digimon yg memiliki Rata-rata HP tertinggi.

```python
df.groupby("Stage")["HP"].mean().sort_values(ascending=False).round(0).head(3).reset_index()
```

9. Lakukan Pengecekan Digimon yg memiliki HP Outliers, Jika ada, tampilkan Nama dan Atk digimon-digimon 
tersebut.

```python
Q1 = df.HP.quantile(0.25)
Q3 = df.HP.quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_HP_outliers = df[(df.HP < lower_bound) | (df.HP > upper_bound)]

df_HP_outliers
```

```python
print(lower_bound, upper_bound, df.HP.min(), df.HP.max())
```

10. Pisahkan Data digimon yg memiliki Atk Outliers dan yg bukan Outliers, sehingga menjadi 2 Dataframe 
baru.

```python
Q1 = df.Atk.quantile(0.25)
Q3 = df.Atk.quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_Atk_outliers = df[(df.Atk < lower_bound) | (df.Atk > upper_bound)]

df_Atk_non_outliers = df[(df.Atk >= lower_bound) & (df.Atk <= upper_bound)]
```

```python
display(df_Atk_outliers)
display(df_Atk_non_outliers)
```
